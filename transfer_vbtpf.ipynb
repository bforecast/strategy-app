```python
import vectorbt as vbt
import sqlite3
import streamlit as st
import pandas as pd
```

```python
connection = sqlite3.connect(st.secrets['sqlite3']['dbname'], check_same_thread=False)
cursor = connection.cursor()
```

```python
df = pd.read_sql("SELECT * FROM portfolio", connection)
df
```

```python
for i in range(len(df)):
        filename = df.loc[i, 'filename']
        id = int(df.loc[i, 'id'])
        with open('portfolio/'+filename, 'rb') as pf_file:
            print(type(id), id, filename)
            pf_blob = pf_file.read()
            try:
                r_set = cursor.execute("UPDATE portfolio SET vbtpf=?  WHERE id=?", (pf_blob, id))
                connection.commit()

                pf_file.close()
            except Exception  as e:
                print("Update portfolio error:", e)
                connection.rollback()
                print(f"Fail to update portfolio('{df.loc[i, 'name']}')")
            else:
                print("Number of records Update:", r_set.rowcount)
```
